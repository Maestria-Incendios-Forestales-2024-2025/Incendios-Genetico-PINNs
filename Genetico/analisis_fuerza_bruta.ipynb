```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt

# Ruta a la carpeta que contiene los CSVs
carpeta = "resultados_fuerza_bruta"

# Lista todos los archivos CSV en la carpeta
csv_files = glob.glob(os.path.join(carpeta, "*.csv"))

# Cargá todos los CSVs en una lista de DataFrames
dataframes = [pd.read_csv(f) for f in csv_files]

# Opcional: combiná todos los DataFrames en uno solo
df_completo = pd.concat(dataframes, ignore_index=True)
```

```python
print("Número total de filas en el DataFrame combinado:", len(df_completo))
print(df_completo.head())
```

```python
df_ordenado = df_completo.sort_values(by='fitness')
mejores_10000 = df_ordenado.head(10000)
```

```python
fig, ax = plt.subplots(figsize=(6,6))

ax.hist(mejores_10000['D'], bins=30, alpha=0.7, edgecolor='black')
ax.axvline(mejores_10000['D'].mean(), color='red', linestyle='dashed', linewidth=2, label=f'Media = {mejores_10000["D"].mean():.2f}')
ax.axvline(50, color='blue', linestyle='dashed', linewidth=2, label='D = 50 (valor de referencia)')
ax.set_xlabel('D', fontsize=14)
ax.set_ylabel('Frecuencia', fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=12)
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('histograma_D.png', dpi=300, bbox_inches='tight')  # Guardar la figura como PNG
plt.show()
```
